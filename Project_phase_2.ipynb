## Importing the relevant libraries and dataset

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("clean_dataset.csv")
```

## Converting the time_series column into a Datetime Object

```python
data["start_time"] = pd.to_datetime(data["start_time"])

data["end_time"] = pd.to_datetime(data["end_time"])
```

```python
data.head()
```

### Convert all columns to numbers:

```python
from sklearn.preprocessing import LabelEncoder

ge_le = LabelEncoder()

data["start_station_name"] = ge_le.fit_transform(data["start_station_name"])


data["end_station_name"] = ge_le.fit_transform(data["end_station_name"])

data["user_type"] = ge_le.fit_transform(data["user_type"])

data["member_gender"] = ge_le.fit_transform(data["member_gender"])

data["bike_share_for_all_trip"] = ge_le.fit_transform(data["bike_share_for_all_trip"])
```

```python
data.info()
```

```python
#Creating Input and Output. This is done using the .iloc[] method.

#X represents the input values of: .

#Y represents the output value of bike_id.

#(Note: We are still dealing with only the Training set)

X = data.iloc[:,0:-5]

Y = data.iloc[:, -5]
```

```python
X
```

```python
#Importing the train_test_split module
from sklearn.model_selection import train_test_split

#Applying the train_test_split to split the data
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.30, random_state = 40)
```

```python
x_train
```

```python
#Import the machine learning algorithm class
from sklearn.linear_model import LinearRegression

#Create an object of the machine learning algorithm class
lr = LinearRegression()

#Fit the machine learning algorithm on the data
lr.fit(x_train.iloc[:, -5:], y_train)

#Score how well the machine learning algorithm did on the training set
score = lr.score(x_train.iloc[:, -5:], y_train)

#Make a prediction of the test set using the trained machine learning algorithm
y_pred = lr.predict(x_test.iloc[:, -5:])
```
